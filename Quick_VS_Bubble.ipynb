```python
import time, random #Importación de librerías

float_list = [random.uniform(1.0, 1000.0) for _ in range(500)] #Lista de numeros flotantes con 500 elementos

#print(float_list)

combined_list_1 = [random.randint(1,1000) for _ in range(500)]
combined_list_1 += [random.uniform(1.0, 1000.0) for _ in range(500)] #Lista combinada con 1,000 elementos

#print(combined_list_1)

combined_list_2 = [random.randint(1,10000) for _ in range(5000)]
combined_list_2 += [random.uniform(1.0, 10000.0) for _ in range(5000)] #Lista combinada con 10,000 elementos

#print(combined_list_2)
```

```python
#BubbleSort

def BubbleSort(l):
  lenght = len(l) #Se obtiene la longuitud de la lista

  for i in range(lenght): #Recorre los elementos de la lista
    #Realiza la comparación del elemento actual con el siguiente hasta el último no ordenado
    for j in range(0, lenght-i-1):
      #Se comparan los elementos adyacentes y se hace el intercambio si es necesario
      if l[j]>l[j+1]:
        l[j], l[j+1] = l[j+1], l[j]


#Asignación de listas para este caso
bl1 = float_list
bl2 = combined_list_1
bl3 = combined_list_2


print(f'Lista sin ordenar: {bl3}')

start = time.time() #Inicialzación del tiempo

BubbleSort(bl3) #Se manda llamar a la función con la lista que deseamos ordenar

end = time.time() #Finalización del tiempo

print(f'El proceso demoró: {end-start}') #Impresión del tiempo de ejecucion

print(f'Lista ordenada: {bl3}')

```
